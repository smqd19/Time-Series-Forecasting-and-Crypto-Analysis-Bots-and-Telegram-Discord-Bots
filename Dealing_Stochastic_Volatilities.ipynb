```python
from statsmodels.tsa.stattools import adfuller
def ad_test(dataset):
     dftest = adfuller(dataset, autolag = 'AIC')
     print("1. ADF : ",dftest[0])
     print("2. P-Value : ", dftest[1])
     print("3. Num Of Lags : ", dftest[2])
     print("4. Num Of Observations Used For ADF Regression:",      dftest[3])
     print("5. Critical Values :")
     for key, val in dftest[4].items():
         print("\t",key, ": ", val)
ad_test(df2['close'])
```

```python
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
test_stationarity(df2)
```

```python
def difference(dataset, interval=1):
    index = list(dataset.index)
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset["close"][i] - dataset["close"][i - interval]
        diff.append(value)
    return (diff)
```

```python
def unlist(list):
    words=''
    for item in list:
        words+=item+' '
    return words
```

```python
def compute_vader_scores(df, label):
    sid = SentimentIntensityAnalyzer()
    df["bearish"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["neg"]) #vader_neg #bearish #negative
    df["neutral"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["neu"]) #vader_neu #nocomment #neutral
    df["bullish"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["pos"]) #vader_pos #bullish #positive
    df["nocomment"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["compound"]) #vader_comp #mixed #can't tell'''
    df['cleantext'] = df[label].apply(lambda x: unlist(x))
    return df
```
